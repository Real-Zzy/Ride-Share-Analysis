```python
def linear_fit(x, y):
    a = np.sum(np.multiply(x,y)) - (np.mean(x) * np.sum(y))
    w = a / (np.sum(np.square(x)) - (np.mean(x) * np.sum(x)))
    b = np.mean(y) - w * np.mean(x)
    return w, b

def linear_predict(w, b, x):
    y = w * x + b
    return y
```

```python
uber_rain = uber.precipIntensity[uber['precipIntensity'] > 0]
uber_rain_price_per_unit = uber.price[uber['precipIntensity'] > 0]/uber.distance[uber['precipIntensity'] > 0]

lyft_rain = lyft.precipIntensity[lyft['precipIntensity'] > 0]
lyft_rain_price_per_unit = lyft.price[lyft['precipIntensity'] > 0]/lyft.distance[lyft['precipIntensity'] > 0]
```

```python
from sklearn.model_selection import train_test_split
Uber_rain_train, Uber_rain_test, Uber_price_train, Uber_price_test = train_test_split(uber_rain, uber_rain_price_per_unit, test_size=0.2, random_state=10)
Lyft_rain_train, Lyft_rain_test, Lyft_price_train, Lyft_price_test = train_test_split(lyft_rain, lyft_rain_price_per_unit, test_size=0.2, random_state=10)


w_U, b_U = linear_fit(Uber_rain_train, Uber_price_train)
w_L, b_L = linear_fit(Lyft_rain_train, Lyft_price_train)
print(f'w_U = {w_U}, b_U = {b_U}')
print(f'w_L = {w_L}, b_L = {b_L}')
```

```python
y_U = linear_predict(w_U, b_U, Uber_rain_train)
y_L = linear_predict(w_L, b_L, Lyft_rain_train)

plt.plot(Uber_rain_train, y_U, color="orange", label = "Fitted Linear Model of Uber")
plt.plot(Lyft_rain_train, y_L, color="green", label = "Fitted Linear Model of Lyft")
plt.title('Price respect to precipitation intensity')
plt.xlabel("Precipitation Intensity")
plt.ylabel("Price per Unit")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
y_U_test = linear_predict(w_U, b_U, Uber_rain_test)
rmse_U = np.sqrt(mean_squared_error(y_U_test, Uber_price_test)) / 2
y_L_test = linear_predict(w_L, b_L, Lyft_rain_test)
rmse_L = np.sqrt(mean_squared_error(y_L_test, Lyft_price_test)) / 2
print(f'The RMSE of the fitted model of Uber = {rmse_U}')
print(f'The RMSE of the fitted model of Lyft = {rmse_L}')
```
